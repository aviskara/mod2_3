```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

heartData = pd.read_csv('heart.csv')
X = heartData.drop(columns=['target'])
y = heartData['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)

avgAccuracy = 0
for x in range(100):
    avgAccuracy += accuracy_score(y_test, predictions)
avgAccuracy /= 100
avgAccuracy

heartData
```

```python
print("This program will predict weather a patient has heart diease with an average of 75% accuracy using 13 paramenters")
age = int(input("Please enter the patient age: "))
sex = int(input("Please enter the patient sex (1 for male, 0 for female): "))
cp = int(input("Please enter the patient chest pain type: "))
trestbps = int(input("Please enter the patient resting blood pressure in mmHg: "))
chol = int(input("Please enter the patient serum cholestoral in mg/dl: "))
fbs = int(input("Please enter the patient fasting blood sugar level: "))
restecg = int(input("Please enter the patient resting electrocardiographic results: "))
thalach = int(input("Please enter the patient maximum heart rate achieved: "))
exang = int(input("Please enter the patient has had exercise induced angina: "))
oldpeak = float(input("Please enter the patient ST depression induced by exercise: "))
slope = int(input("Please enter the patient the slope of the peak exercise ST segment: "))
ca = int(input("Please enter the patient number of major vessels (0-3) colored by flourosopy: "))
thal = int(input("Please enter the patient thal: 3 = normal; 6 = fixed defect; 7 = reversable defect: "))

prediction = model.predict([ [age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal] ])
if prediction[0] == 1:
    print("The patient likely has a heart diease")
else:
    print("The patient does not likely have a heart diease")
```
